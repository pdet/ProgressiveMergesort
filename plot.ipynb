```python
import pandas as pd
import matplotlib.pyplot as plt

class Experiment:
    num_queries = 1000
    column_size = 1000000
    append_size = 10
    append_frequency = 10
    start_updates_after = 50
    delta = 0.1;
    file_name = 'results/'

    def __init__(self, num_queries, column_size,append_size,append_frequency,start_updates_after,delta):
        self.num_queries = num_queries
        self.column_size = column_size
        self.append_size = append_size
        self.append_frequency = append_frequency
        self.start_updates_after = start_updates_after
        self.delta = delta
        self.file_name = 'results/' +str(num_queries) + '_' + str(column_size) + '_' \
        + str(append_size) + '_' + str(append_frequency) + '_' + str(start_updates_after) + '_' + str(delta)

    def per_query_plot(self):
        complete = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(1) +\
                               ".csv",delimiter=';', names=['total'])
        gradual = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(2) +\
                               ".csv",delimiter=';', names=['total'])
        ripple = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(3) +\
                               ".csv",delimiter=';', names=['total'])
        mergesort = pd.read_csv(self.file_name+"_" + str(2)+"_" + str(4) +\
                               ".csv",delimiter=';', names=['total'])
        plt.plot(complete['total'], label='MC')
        plt.plot(gradual['total'], label='MG')
        plt.plot(ripple['total'], label='MR')
        plt.plot(mergesort['total'], label='PI')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',)
        plt.yscale("log")
        plt.xlabel('Query #')
        plt.ylabel('Time (s)')
        plt.show()
    
    def pi(self):
        mergesort = pd.read_csv(self.file_name+"_" + str(2)+"_" + str(4) +\
                                ".csv",delimiter=';', names=['total'])
        
        name = 'results/' +str(self.num_queries) + '_' + str(self.column_size) + '_' \
        + str(self.append_size*10) + '_' + str(self.append_frequency) + '_' +\
        str(self.start_updates_after) + '_' + str(self.delta)
        
        mergesort_10 = pd.read_csv(name+"_" + str(2)+"_" + str(4) +\
                                ".csv",delimiter=';', names=['total'])
        
        name = 'results/' +str(self.num_queries) + '_' + str(self.column_size) + '_' \
        + str(self.append_size*100) + '_' + str(self.append_frequency) + '_' +\
        str(self.start_updates_after) + '_' + str(self.delta)
        
        mergesort_100 = pd.read_csv(name+"_" + str(2)+"_" + str(4) +\
                                ".csv",delimiter=';', names=['total'])
        plt.plot(mergesort['total'][:190], label='PI - 100')
        plt.plot(mergesort_10['total'][:190], label='PI - 1000')
        plt.plot(mergesort_100['total'][:190], label='PI - 10000')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',)
        plt.yscale("log")
        plt.xlabel('Query #')
        plt.ylabel('Time (s)')

        plt.show()
    
    def cum_plot(self):
        complete = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(1) +\
                               ".csv",delimiter=';', names=['total'])
        gradual = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(2) +\
                               ".csv",delimiter=';', names=['total'])
        ripple = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(3) +\
                               ".csv",delimiter=';', names=['total'])
        mergesort = pd.read_csv(self.file_name+"_" + str(2)+"_" + str(4) +\
                               ".csv",delimiter=';', names=['total'])
        plt.plot(complete['total'].cumsum(), label='MC')
        plt.plot(gradual['total'].cumsum(), label='MG')
        plt.plot(ripple['total'].cumsum(), label='MR')
        plt.plot(mergesort['total'].cumsum(), label='PI')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',)
        plt.yscale("log")
        plt.xlabel('Query #')
        plt.ylabel('Time (s)')
        plt.show()

    def show_sum(self):
        complete = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(1) +\
                               ".csv",delimiter=';', names=['total'])
        gradual = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(2) +\
                               ".csv",delimiter=';', names=['total'])
        ripple = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(3) +\
                               ".csv",delimiter=';', names=['total'])
        mergesort = pd.read_csv(self.file_name+"_" + str(2)+"_" + str(4) +\
                               ".csv",delimiter=';', names=['total'])
        print("MC: " + str(complete[1000:].sum()))
        print("MG: " + str(gradual[1000:].sum()))
        print("MR: " + str(ripple[1000:].sum()))
        print("PM: " + str(mergesort[1000:].sum()))
        
    def show_count(self):
        complete = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(1) +\
                               ".csv",delimiter=';', names=['total'])
        gradual = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(2) +\
                               ".csv",delimiter=';', names=['total'])
        ripple = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(3) +\
                               ".csv",delimiter=';', names=['total'])
        mergesort = pd.read_csv(self.file_name+"_" + str(2)+"_" + str(4) +\
                               ".csv",delimiter=';', names=['total'])
        print("MC: " + str(complete[1000:].count()))
        print("MG: " + str(gradual[1000:].count()))
        print("MR: " + str(ripple[1000:].count()))
        print("PM: " + str(mergesort[1000:].count()))
        
    def show_robustness(self):
        complete = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(1) +\
                               ".csv",delimiter=';', names=['total'])
        gradual = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(2) +\
                               ".csv",delimiter=';', names=['total'])
        ripple = pd.read_csv(self.file_name+"_" + str(1)+"_" + str(3) +\
                               ".csv",delimiter=';', names=['total'])
        mergesort = pd.read_csv(self.file_name+"_" + str(2)+"_" + str(4) +\
                               ".csv",delimiter=';', names=['total'])
        print("MC: " + str(complete[1000:].var()))
        print("MG: " + str(gradual[1000:].var()))
        print("MR: " + str(ripple[1000:].var()))
        print("PM: " + str(mergesort[1000:].var()))
              
    def plot(self):
        print(self.file_name)
        self.per_query_plot()
        self.cum_plot()
```
